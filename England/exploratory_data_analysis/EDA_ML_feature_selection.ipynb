```python
ml_data = ml_data.drop('id',1)
label = ml_data['home_win'].values
features = ml_data.drop('home_win',1)
features.head()
```

## Pearson Correlation

```python
pearson_corr = ml_data.corr()
plt.subplots(figsize=(20,10))
sns.heatmap(pearson_corr)
plt.show()
```

```python
home_win_pears_corr = pearson_corr['home_win'].drop('home_win')
home_win_pears_corr = np.abs(home_win_pears_corr)
plt.subplots(figsize=(20,10))
home_win_pears_corr.plot(kind='bar')
plt.title('Pearson correlation (absolute value) between features and output var')
plt.show()
```

```python
selec_pearson_feat = home_win_pears_corr[home_win_pears_corr>0.1]
list(selec_pearson_feat.index.values)
```

## Kendall Correlation

```python
kendall_corr = ml_data.corr('kendall')
plt.subplots(figsize=(20,10))
sns.heatmap(kendall_corr)
plt.show()
```

```python
home_win_kendall_corr = kendall_corr['home_win'].drop('home_win')
home_win_kendall_corr = np.abs(home_win_kendall_corr)
plt.subplots(figsize=(20,10))
home_win_kendall_corr.plot(kind='bar')
plt.title('Kendall correlation (absolute value) between features and output var')
plt.show()
```

```python
selec_kendal_feat = home_win_kendall_corr[home_win_kendall_corr>0.1]
list(selec_kendal_feat.index.values)
```

## Point Biserial correlation with ouptut var

```python
from scipy.stats import pointbiserialr

output_var = 'home_win'
output = ml_data[output_var].values
feat_names = ml_data.drop(output_var,1).columns.values

threshold_r = 0.1
selected_feat = []
for feat in feat_names:
    r,p_value = pointbiserialr(output, ml_data[feat].values)
    if np.abs(r)>threshold_r:
        selected_feat.append(feat)
        print(feat)
        print('Coeff: %f' % r)
        print(p_value)
        print('')
        
print(selected_feat)
print(len(selected_feat))
    
```

## Random Forest Feat Importance Score

```python
from sklearn.ensemble import RandomForestClassifier
rdmF = RandomForestClassifier(n_estimators=500)

output_var='home_win'
y=ml_data[output_var].values
X = ml_data.drop(output_var,1).values
feat_names = ml_data.drop(output_var,1).columns.values
rdmF.fit(X, y)

importances = rdmF.feature_importances_
print('Importance scores')
print(importances)
fig,ax=plt.subplots(figsize=(20,10))
feat_importance_df = pd.DataFrame({'features': feat_names, 'importance': importances})
feat_importance_df.plot(x='features', y='importance', kind='bar', rot=90,ax=ax)
title = 'Importance score of each feature' 
plt.title(title)
plt.show()
```

```python
selec_feat_rdmf = [feat_names[i] for i in range(len(importances)) if importances[i]>(1.0/len(feat_names))]
selec_feat_rdmf
```
