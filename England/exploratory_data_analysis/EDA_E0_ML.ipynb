# EDA 

## Load Python modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Data

```python
filepath = '../data/ML/E0_ML.csv'
data = pd.read_csv(filepath)
data[:50]
```

## Type of data

```python
data.dtypes
```

## Is the data balanced regarding the output variable?

```python
OUTPUT_VAR = 'home_win'

y = data[OUTPUT_VAR].values
classes = np.unique(y)

for cls in classes:
    size_cls = len([y_i for y_i in y if y_i==cls])
    print('Size of class %d: %d' % (cls, size_cls))
```

```python
data.describe()
```

## More advanced statistics

### Univariate analysis

```python
features = data.columns
predictors = [x for x in features if x!=OUTPUT_VAR]

TYPES_FIG = ['Histogram', 'Boxplot', 'Violinplot']

nb_predictors = len(predictors)
nb_fig_per_row = len(TYPES_FIG)
nb_rows = nb_predictors

plt.figure(figsize=(20,20))
for i in range(nb_predictors):
    predictor = predictors[i]
    predictor_values = data[predictor].values
    plt.subplot(nb_rows,nb_fig_per_row,i*nb_fig_per_row+1)
    sns.distplot(predictor_values)
    tit1 = TYPES_FIG[0] + ': ' + predictor
    plt.title(tit1)

    plt.subplot(nb_rows,nb_fig_per_row,i*nb_fig_per_row+2)
    sns.boxplot(predictor_values)
    tit2 = TYPES_FIG[1] + ': ' + predictor
    plt.title(tit2)

    plt.subplot(nb_rows,nb_fig_per_row,i*nb_fig_per_row+3)
    sns.violinplot(predictor_values)
    tit3 = TYPES_FIG[2] + ': ' + predictor
    plt.title(tit3)
plt.show()
```

#### Skewness of the features

```python
from scipy.stats import skew

for predictor in predictors:
    predictor_values = data[predictor].values
    skewness = skew(predictor_values)
    skewness_info = 'The skewness for %s is %f' % (predictor, skewness)
    print(skewness_info)
```

#### QQ-Plots for normality test

```python
from statsmodels.graphics.gofplots import qqplot

for predictor in predictors:
    predictor_values = data[predictor].values
    qqplot(predictor_values,line='s')
    tit = 'QQ-Plot normal: ' + predictor
    plt.title(tit)
plt.show()
```

### Bivariate analysis

#### Scatter plot: correlations between non_categorical predictors

```python
non_categorical_data = data.select_dtypes(include=[np.float])
non_categorical_predictors = [x for x in non_categorical_data.columns if x!=OUTPUT_VAR]

from itertools import combinations

two_by_two_predictors = combinations(non_categorical_predictors, 2)

for feat1, feat2 in two_by_two_predictors:
    feat1_values = non_categorical_data[feat1].values
    feat2_values = non_categorical_data[feat2].values
    plt.figure()
    plt.scatter(feat1_values, feat2_values)
    plt.xlabel(feat1)
    plt.ylabel(feat2)
plt.show()
```
